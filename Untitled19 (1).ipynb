```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
from scipy.interpolate import CubicSpline

# Voronoi Field Calculation
def voronoi_field(grid, obstacles, weight=1.0, alpha=1.0, dmax=10):
    if len(obstacles) < 4:
        raise ValueError("At least 4 distinct obstacle points are required for Voronoi diagram.")

    vor = Voronoi(obstacles)
    field = np.zeros_like(grid, dtype=float)

    for x in range(grid.shape[0]):
        for y in range(grid.shape[1]):
            d0 = np.min(np.sqrt((obstacles[:, 0] - x)**2 + (obstacles[:, 1] - y)**2))
            dv = np.min(np.sqrt((vor.vertices[:, 0] - x)**2 + (vor.vertices[:, 1] - y)**2))
            if d0 < dmax:
                field[x, y] = weight * (alpha / (alpha + d0)) * (dv / (d0 + dv)) * ((d0 - dmax) / dmax)**2
    return field

# Simplified Theta-star Algorithm
def theta_star(grid, start, goal, voronoi_field):
    # Placeholder for actual Theta-star implementation
    path = [start, (10, 10), (20, 20), (30, 25), goal]
    return path

# Simple Cubic Spline Smoothing
def cubic_spline_smoothing(path):
    path = np.array(path)
    t = np.arange(len(path))
    cs_x = CubicSpline(t, path[:, 0])
    cs_y = CubicSpline(t, path[:, 1])

    t_new = np.linspace(0, len(path) - 1, num=100)
    smooth_path = np.vstack((cs_x(t_new), cs_y(t_new))).T
    return smooth_path

# Main Function
def main():
    # Define grid and obstacles
    grid_size = (50, 30)
    grid = np.zeros(grid_size)
    obstacles = np.array([[10, 10], [15, 15], [20, 20], [25, 25], [30, 30], [10, 20], [20, 10]])

    # Calculate Voronoi field
    vor_field = voronoi_field(grid, obstacles)

    # Define start and goal positions
    start = (0, 0)
    goal = (49, 29)

    # Run Theta-star algorithm
    path = theta_star(grid, start, goal, vor_field)

    # Smooth path with cubic spline
    smoothed_path = cubic_spline_smoothing(path)

    # Plot results
    plt.imshow(grid.T, cmap='gray_r')
    voronoi_plot_2d(Voronoi(obstacles), ax=plt.gca(), show_vertices=False, line_colors='orange')
    plt.plot(*zip(*path), marker='o', color='blue', label='Theta* Path')
    plt.plot(*zip(*smoothed_path), marker='x', color='red', label='Cubic Spline Smoothed Path')
    plt.scatter(*zip(*obstacles), color='black', label='Obstacles')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
